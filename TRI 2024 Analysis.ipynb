```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r'C:\Users\olibh\OneDrive\Desktop\TRI US 2024.xlsx')
```

```python
## Let's clean the data and select only the Defense contractors
```

```python
companies = ["BAE", "LOCKHEED", "NORTHROP"]

df_filtered = df[
    df["17. STANDARD PARENT CO NAME"]
    .astype(str)
    .str.upper()
    .str.contains('|'.join(companies), na=False)]
```

```python
df_filtered.head(3)
```

```python
##let's keep the columns that we need 
```

```python
df_clean = df_filtered[[
    "1. YEAR",
    "4. FACILITY NAME",
    "6. CITY",
    "8. ST",
    "17. STANDARD PARENT CO NAME",
    "37. CHEMICAL",
    "51. 5.1 - FUGITIVE AIR",
    "52. 5.2 - STACK AIR"
]]
```

```python
df_clean.head(2)
```

```python
df_clean.shape
```

```python
import warnings
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
df_clean["51. 5.1 - FUGITIVE AIR"] = pd.to_numeric(
    df_clean["51. 5.1 - FUGITIVE AIR"], errors="coerce"
)

df_clean["52. 5.2 - STACK AIR"] = pd.to_numeric(
    df_clean["52. 5.2 - STACK AIR"], errors="coerce"
)
```

```python
df_clean.head(3)
```

```python
df_clean["air_emissions_lbs"] = (
    df_clean["51. 5.1 - FUGITIVE AIR"].fillna(0) +
    df_clean["52. 5.2 - STACK AIR"].fillna(0)
)
```

```python
df_clean = df_clean.drop(columns=[
    "51. 5.1 - FUGITIVE AIR",
    "52. 5.2 - STACK AIR"
])
```

```python
df_clean.columns = [
    "year",
    "facility_name",
    "city",
    "state",
    "parent_company",
    "chemical",
    "air_emissions_lbs"
]
```

```python
df_clean = df_clean[df_clean["air_emissions_lbs"] > 0]
df_clean = df_clean.drop_duplicates()
```
